```python
import requests
```

```python
import pandas as pd
```

```python
url = "https://fapi.binance.com/fapi/v1/exchangeInfo"
```

```python
response = requests.get(url)
```

```python
print(response)
```

```python
futures_exchange_info = response.json()
```
