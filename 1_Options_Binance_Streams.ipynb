```python
import websocket
import json
```

```python
def on_message(ws, message):
    data = json.loads(message)
    print(data)
```

```python
def on_error(ws, error):
    print(f"Error: {error}")
```

```python
def on_close(ws, close_status_code, close_msg):
    print("Closed")
```

# Trade Streams

The Trade Streams push raw trade information for specific symbol or underlying asset.<br>

Stream Name: symbol@trade or underlyingAsset@trade

BTC-200630-9000-P // BTC

Update Speed: 50ms

```python
url = 'wss://nbstream.binance.com/eoptions/ws/'
```

```python
symbol = 'BTC'
```

```python
stream = '@trade'
```

```python
long_url = url+symbol+stream
```

```python
print(long_url)
```

```python
ws = websocket.WebSocketApp(long_url, on_message=on_message, on_error=on_error, on_close=on_close)
```

```python
ws.run_forever()
```

### Payload

{<br>
    "e":"trade",                        // event type   <br>
    "E":1591677941092,                  // event time   <br>
    "s":"BTC-200630-9000-P",            // Option trading symbol   <br>
    "t":1,                              // trade ID   <br>
    "p":"1000",                         // price   <br>
    "q":"-2",                           // quantity   <br>
    "b":4611781675939004417,            // buy order ID   <br>
    "a":4611781675939004418,            // sell order ID   <br>
    "T":1591677567872,                  // trade completed time<br>  
    "S":"-1"                            // direction, -1 for taker sell, 1 for taker buy   <br>
}<br>

# Kline / Candlestick Streams

The Kline/Candlestick Stream push updates to the current klines/candlestick every 1000 milliseconds (if existing).<br>

Kline/Candlestick chart intervals:<br>

m -> minutes; h -> hours; d -> days; w -> weeks; M -> months<br>

"1m", "3m", "5m", "15m" "30m" "1h", "2h", "4h", "6h", "12h", "1d", "3d", "1w",<br>

Stream Name:symbol@kline_interval <br>

Update Speed: 1000ms<br>

```python
url = 'wss://nbstream.binance.com/eoptions/ws/'
```

```python
symbol = 'BTC-240628-15000-C'
```

```python
stream = '@kline_'
```

```python
interval = '1m'
```

```python
long_url = url + symbol + stream + interval
```

```python
print(long_url)
```

```python
ws = websocket.WebSocketApp(long_url, on_message=on_message, on_error=on_error, on_close=on_close)
```

```python
ws.run_forever()
```

### Payload

{<br>
    "e":"kline",                        // event type   <br>
    "E":1638747660000,                  // event time   <br>
    "s":"BTC-200630-9000-P",            // Option trading symbol   <br>
    "k":{                             <br>
        "t":1638747660000,              // kline start time   <br>
        "T":1638747719999,              // kline end time  <br>
        "s":"BTC-200630-9000-P",        // Option trading symbol   <br>

        "x":false,                      // current candle has been completed Y/N   <br>

        "V":"0",                        // taker completed trade volume (in contracts)     <br>        
        "Q":"0"                         // taker trade amount(in quote asset)   <br>
    }<br>

```python
url = 'wss://nbstream.binance.com/eoptions/ws/'
```

```python
symbol = 'BTC-240414-64000-C'
```

```python
stream = '@depth'
```

```python
levels = '10'
```

```python
speed = '@100ms'
```

```python
long_url = url + symbol + stream + levels + speed
```

```python
print(long_url)
```

```python
ws = websocket.WebSocketApp(long_url, on_message=on_message, on_error=on_error, on_close=on_close)
```

```python
ws.run_forever()
```
